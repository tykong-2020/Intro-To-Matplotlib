```python
%matplotlib inline
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
```

```python
x = np.linspace(0, 2*np.pi, 100)           # gives a set of 100 points from 0 to 2 pi
sin_x = np.sin(x)
```

```python
figure, ax = plt.subplots()
_ = ax.plot(x, sin_x)
```

```python
figure, ax = plt.subplots(figsize = (15, 5))
_ = ax.plot(x, sin_x, marker = '*', markerfacecolor = 'yellow', markeredgecolor = 'red', 
            linestyle = "", color = 'r')
_ = ax.set_aspect('equal')
```

```python
fig, ax = plt.subplots()
_ = ax.plot(x, sin_x, markersize=20, linewidth=20)
_ = ax.set_aspect('equal')
```

```python
fig, ax = plt.subplots(figsize = (15, 5))
_ = ax.plot(x, sin_x, markersize=20, linewidth=20, markeredgewidth = 1, markeredgecolor = 'cyan')
_ = ax.set_aspect('equal')
```

```python
fig, ax = plt.subplots()
_ = ax.plot(x, sin_x, markersize = 20, linestyle = '-.')
_ = ax.set_aspect('equal')
```

```python
fig, ax = plt.subplots(figsize = (15, 5))
_ = ax.plot(x, sin_x, drawstyle = 'steps')
_ = ax.set_aspect('equal')
```

```python
fig, ax = plt.subplots(figsize = (15, 5))
_ = ax.plot(x, sin_x, drawstyle = 'steps', label = 'sin')
_ = ax.set_aspect('equal')
_ = ax.legend()
```

```python
fig, ax = plt.subplots(figsize = (15, 5))
_ = ax.plot(x, sin_x, drawstyle = 'steps', label = 'sin')
_ = ax.plot(x, x, label = 'x')
_ = ax.set_aspect('equal')
_ = ax.legend()
```

```python
cos_x = np.cos(x)
```

```python
fig, ax = plt.subplots(figsize = (20, 5))
_ = ax.plot(x, sin_x, drawstyle = 'steps', label = 'sin')
_ = ax.plot(x, x, label = 'x')
_ = ax.plot(x, cos_x, label = 'cos')
_ = ax.set_aspect('equal')
_ = ax.legend(loc = 0)
```
